```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x = []
with open("csv_files/test_y_continuous.csv") as f:
    for row in f:
        i = float(row.split(",")[0])
        x.append(np.array([i]))
x = np.array(x)
for i in range(7):
    y = []
    with open("csv_files/test_x.csv") as f:
        for row in f:
            a = float(row.split(",")[i])
            y.append(np.array([a]))
    y = np.array(y)
    plt.figure(figsize=(16, 12))
    linear_fit = LinearRegression().fit(x, y)
    r2 = linear_fit.score(x, y)
    y_pred = linear_fit.predict(x)
    plt.title(f"Feature {i} - r2: {r2}")
    plt.scatter(x, y, color="black")
    plt.plot(x, y_pred, color="blue", linewidth=3)
    plt.xlabel("x")
    plt.ylabel(f"feature {i}")
    plt.savefig(f'features/feature_{i}.pdf')
    plt.show()
    
```
